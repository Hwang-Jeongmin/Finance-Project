```python
# 제일 늦은 것은 2016-03-03 Nikkey225
# 따라서 2016-03-03부터의 데이터를 사용하기로 한다.
# 또한 수익률을 측정해야하기 때문에 pct_change()를 사용해 수익률에 해댱하는 열을 만들어줘야한다.
for file in ohlcv_file_list:
  name = file.split('.')[0]
  df = pd.read_csv(f'ohlcv_data/{file}')
  start_date = df.Date[0]
  print(name, ": ", start_date)
```

```python
def import_ohlcv_data(start, file):
  file_path = f'ohlcv_data/{file}'  # 파일 불러오기
  df = pd.read_csv(file_path)
  df = df[df.Date >= start].reset_index(drop=True)  # 시작하는 날짜보다 뒤에 있는 것만 가져오기
  df["Rate"] = df['Close'].pct_change()
  df.dropna(inplace=True)
  df['Rate'] = (np.exp(np.log(df["Rate"]+1).rolling(5).sum())-1).shift(-5)
  df.dropna(inplace=True)
  
  return df
```

```python
def import_commodity_data(start, file):
  file_path = f'commodity/{file}'  # 파일 불러오기
  name = file.split('.')[0]
  df = pd.read_csv(file_path)
  df = df[df.Date >= start].reset_index(drop=True)  # 시작하는 날짜보다 뒤에 있는 것만 가져오기
  df = df.iloc[:,:2]                                # Date를 뺀 가장 첫번째 열이 모두 가격을 나타내므로 이것만을 사용하기로 한다
  df.columns = ['Date', f'{name} Price']                    # 열 이름을 똑같이 변경
  return df
```

```python
def merge_data_set(dataset, feature_list):
  df = dataset
  for feature in feature_list:
    df = pd.merge(df, feature, how='left')
  df[['USD Change', 'JPY Change', 'CNY Change', 'EUR Change']] = df[['USD Change', 'JPY Change', 'CNY Change', 'EUR Change']].fillna(0)
  df = df.fillna(method='ffill')
  return df
```

### KODEX200

```python
feature_list = total_feature_list
KODEX200_combined = merge_data_set(KODEX200, feature_list)
KODEX200_combined.head()
```

```python
KODEX200_combined.drop(['Date', 'KODEX200_Close'], axis=1, inplace=True)
```

```python
def Rate_into_labels(dataset):
  q1 = round(dataset["Rate"].quantile(0.2),4)
  q2 = round(dataset["Rate"].quantile(0.4),4)
  q3 = round(dataset["Rate"].quantile(0.6),4)
  q4 = round(dataset["Rate"].quantile(0.8),4)
  dataset['Rate'] = pd.qcut(dataset["Rate"],5, labels=[0,1,2,3,4])
  quantile_df = pd.DataFrame({'Rate':[q1,q2,q3,q4]}, index=['20%', '40%', '60%', '80%'])
  quantile_df2 = pd.DataFrame([f'~{q1}',f'{q1}~{q2}', f'{q2}~{q3}', f'{q3}~{q4}', f'{q4}~'], index=['~20%', '20% ~ 40%', '40% ~ 60%', '60% ~ 80%', '80%~'])
  return quantile_df2
```

```python
Rate_into_labels(KODEX200_combined)
```

```python
KODEX200_combined.Rate.value_counts()
```

```python
X = KODEX200_combined.drop('Rate', axis=1)
y = KODEX200_combined[['Rate']]
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X_scaled,y, test_size=0.2, shuffle=False)
```

```python
X_train = pd.DataFrame(X_train, columns = KODEX200_combined.columns.drop("Rate"))
X_test = pd.DataFrame(X_test, columns = KODEX200_combined.columns.drop("Rate"))
```

```python
np.random.seed(42)
from sklearn.feature_selection import RFE
from lightgbm import LGBMClassifier
estimator = LGBMClassifier(boosting_type='gbdt')
selector = RFE(estimator, n_features_to_select=15, step = 3)
selector.fit(X_train, y_train)
```

```python
selected_column_list = X_train.columns[selector.support_]
print("Selected columns")
pd.DataFrame({'selected_columns':X_train.columns[selector.support_]}).head()
```

```python
pd.DataFrame({'columns':X_train.columns, 'Ranking': selector.ranking_}).sort_values(by='Ranking', ascending=True).reset_index(drop=True).head()
```

```python
X_train = X_train[selected_column_list]
X_test = X_test[selected_column_list]
X_train.head()
```

#### LightGBM

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
lgbm = LGBMClassifier()
lgbm.fit(X_train, y_train)
```

```python
pred = lgbm.predict(X_test)
print(f"accuracy: {round(accuracy_score(pred, y_test),4)}")
print(f"recall: {round(recall_score(pred, y_test, average='micro'),4)}")
print(f"precision: {round(precision_score(pred, y_test, average='micro'),4)}")
print(f"f1_score: {round(f1_score(pred, y_test, average='micro'),4)}")
```

```python
confusion_matrix(pred, y_test)
```
