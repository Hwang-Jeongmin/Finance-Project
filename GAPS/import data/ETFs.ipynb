```python
df2 = stock.get_etf_price_deviation('20150101', '20220520','069500')
df3 = stock.get_etf_tracking_error('20150101', '20220520','069500')
display(df2.head())
display(df3.head())
```

```python
ticker_list = ['069500', '232080', '143850', '195930', 
            '238720', '192090','148070', '136340',
            '182490', '132030', '130680', '114800',
            '138230', '139660','130730']

for ticker in ticker_list:
    df1 = stock.get_etf_ohlcv_by_date("20150220", "20220520", ticker)
    df1['NAV'] = df1['NAV'].apply(lambda x: x/100)
    df1['등락률'] = df1['종가'].pct_change()
    df1['등락률'].fillna(0, inplace=True)
    
    df2 = stock.get_etf_price_deviation("20190101", "20220520", ticker)
    df3 = stock.get_etf_tracking_error("20190101", "20220520", ticker)
    
    df = pd.merge(df1, df2)
    df = pd.merge(df, df3)
    name = stock.get_etf_ticker_name(ticker)
    df.to_csv(f'{name}.csv', header=True, encoding="utf-8-sig")
```
