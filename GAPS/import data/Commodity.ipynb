#### 원자재 데이터가져오기

```python
# 서부 텍사스 중질유(WTI Crude)
oil = quandl.get('EIA/PET_RWTC_D', start_date='2010-01-01')
oil.tail()
```

```python
# "LBMA/GOLD", 런던 금시장협회 (LBMA) 금 가격
# 각각의 칼럼은 측정된 시점과 통화를 나타낸다.
# GBP(AM)은 오전의 영국 파운드화 금 가격
gold = quandl.get('LBMA/GOLD', start_date='2010-01-01')
gold.tail()
```

```python
# "LBMA/SILVER", 런던 금시장협회 (LBMA) 은 가격
silver = quandl.get('LBMA/SILVER', start_date='2010-01-01')
silver.tail()
```

```python
# Copper; grade A cathode; LME spot price; CIF European ports; US$ per metric ton
copper = quandl.get('ODA/PCOPP_USD', start_date='2010-01-01')
copper.tail()
```

```python
# 천연가스 선물(NYMEX)
natural_gas = fdr.DataReader('NG', start='2010')
natural_gas.tail()
```

```python
oil.to_csv('oil.csv')
gold.to_csv('gold.csv')
silver.to_csv('silver.csv')
copper.to_csv('copper.csv')
natural_gas.to_csv('natural_gas.csv')
```
